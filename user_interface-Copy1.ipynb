```python
class prep:
    @staticmethod
    def prep1():
        global file_path
        img_path = file_path
        img = load_img(img_path, target_size=(150, 150))
        img_tensor = img_to_array(img)                   
        img_tensor = np.expand_dims(img_tensor, axis=0)  
        img_tensor /= 255.                               
        return img_tensor
    @staticmethod
    def prep2():
        global file_path
        img_path = file_path
        img = load_img(img_path, target_size=(160, 160))
        img_tensor = img_to_array(img)                   
        img_tensor = np.expand_dims(img_tensor, axis=0)  
        img_tensor /= 255.                               
        return img_tensor
```

```python
p=prep()
class check:
    
    @staticmethod
    def load_image_1():
        global file_path
        new_model1= tf.keras.models.load_model('chest_oct.h5')
        img_path = file_path
        new_image = p.prep1()
        pred = new_model1.predict(new_image)

        if pred>=0.5:
            return "oct"
        else: 
            return "chest"
    @staticmethod    
    def load_image_2():
        global file_path
        new_model2= tf.keras.models.load_model('pneumonia.h5')
        img_path = file_path
        new_image = p.prep1()
        pred = new_model2.predict(new_image)
        if pred>=0.5:
            return "pneumonia"
        else: 
            return "normal"
    @staticmethod
    def load_image_3():
        global file_path
        new_model3= tf.keras.models.load_model('pneumonia222.h5')
        img_path = file_path
        new_image = p.prep1()
        pred = new_model3.predict(new_image)
        if pred>=0.5:
            return "Diagnosis :  The patient has been detected with bacterial pneumonia.\n\nThere is an infection in the lungs caused due to bacteria.\n\nPatient might notice symptoms like cough, fever, shortness of breath, chest pain, fatigueness over the time.\n\n\n\n Treatment: Doctor probably will prescribe antibiotics.\n\nIt’s very important to finish all of those.\n\nOtherwise the bacteria may not all be killed andpatient could get sick all over again.\n\nDoctor might also suggest medication for pain and fever.\n\nGet lots of rest.\n\nDrink plenty of fluids (they’ll loosen up the gunk in lungs soit can be coughed out).\n\nUse a humidifier or take a warm bath (more gunk-loosening).\n\nDon’t smoke.\n\nStay home until fever goes down and aren’t coughing anything out."
        else: 
            return "Diagnosis :  The patient has been detected with viral pneumonia.\n\nThere is an infection in the lungs caused due to cold and flu viruses.\n\nPatient might notice symptoms like fever, chills, dry cough, muscle pain, headache over the time.\n\n\n\n Treatement: Doctor may give an antiviral medication.\n\nIt’s important to get plenty of rest.\n\nMight take over-the-counter medicines to fight your fever and ease pain.\n\nBe sure to drink plenty of fluids.\n\nCheck in with doctor after treatment to make sure lungs are clear.\n\nIn severe cases, patient may need to stay in the hospital to get help breathing."
    @staticmethod    
    def load_image_4():
        global file_path

        new_model4= tf.keras.models.load_model('oct_model.h5')
        img_path = file_path
        new_image = p.prep2()
        pred = new_model4.predict(new_image)
        max=np.amax(pred)
        result = np.where(pred == np.amax(pred))
        listOfCordinates = list(zip(result[0], result[1]))
        for cord in listOfCordinates:
            if cord[1]==0:
                return "Diagnosis :  The patient has been detected with CNV in eyes.\n\nCNV can create a sudden deterioration of central vision, noticeable within a few weeks.\n\nOther symptoms which can occur include colour disturbances, and metamorphopsia.\n\n\n\nTreatment :  Consult doctor immediately for further treatments which may include laser, pdt or anti - VEGF therapy."
            elif cord[1]==1:
                return "Diagnosis :  The patient has been detected with DME in eyes.\n\nEarlier, there will be no symptoms.\n\nBut later, patient might experience blurry vision, floating spots in vision, blindness.\n\n\n\nTreatment :  Dilated eye exam is suggested.\n\nDoctor might recommend medicine, laser treatment or surgery depending on the condition."
            elif cord[1]==2:
                return "Diagnosis :  The patient has been detected with DRUSEN in eyes.\n\nDrusen are small yellow deposits of fatty proteins (lipids) that accumulate under the retina.\n\nThe patient might experience difficulty adapting from bright lights to low lights, hazy or blurry vision, or blank spot in your central vision.\n\n\n\nTreatment :  Detailed check up is suggested. "
            elif cord[1]==3:
                return "Diagnosis :  The patient's eyes are perfectly normal."
```
